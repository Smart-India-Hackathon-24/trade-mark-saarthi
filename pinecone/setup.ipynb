```python
# Define a function to split the data into batches
def split_into_batches(data, batch_size=100):
    """Splits data into batches of a given size."""
    for i in range(0, len(data), batch_size):
        yield data[i:i + batch_size]

# Function to upsert embeddings in smaller batches
def upsert_in_batches(data):
    """Upserts data in batches to avoid exceeding size limits."""
    for batch in split_into_batches(data, batch_size=100):
        # Extract titles and embeddings from the batch
        titles_batch, embeddings_batch = zip(*batch)
        
        # Convert to the format required by Pinecone: [(id, vector), ...]
        upsert_batch = [(title, embedding) for title, embedding in zip(titles_batch, embeddings_batch)]
        
        # Perform the upsert operation for the batch
        index.upsert(upsert_batch)

# Perform the batch upsert
upsert_in_batches(upsert_data)
```

```python
import pinecone

# Function to split the data into batches
def split_into_batches(data, batch_size=100):
    """Splits data into batches of a given size."""
    for i in range(0, len(data), batch_size):
        yield data[i:i + batch_size]

# Function to upsert embeddings in smaller batches with error handling
def upsert_in_batches(data):
    """Upserts data in batches, skipping problematic IDs."""
    for batch in split_into_batches(data, batch_size=100):
        sanitized_batch = []
        for title, embedding in batch:
            try:
                # Attempt to upsert the vector, if the title is invalid (non-ASCII), it will raise an exception
                index.upsert([(title, embedding)])
                sanitized_batch.append((title, embedding))
            except pinecone.exceptions.PineconeApiException as e:
                # Handle the exception for non-ASCII characters or any other invalid vector IDs
                print(f"Skipping title: {title} due to error: {e}")
                continue  # Skip the problematic title and continue with the next
        print("Done")

# Perform the batch upsert
upsert_in_batches(upsert_data)
```
