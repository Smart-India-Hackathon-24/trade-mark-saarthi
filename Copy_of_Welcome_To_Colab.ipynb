```
from googletrans import Translator
from sentence_transformers import SentenceTransformer, util
from fuzzywuzzy import fuzz
from phonetics import metaphone

# Initialize the translator and embedding model
translator = Translator()
model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

def are_strings_equivalent(string1, L, semantic_threshold=0.8, fuzzy_threshold=85, phonetic_threshold=80):
    """
    Determines if two strings have the same meaning or similar pronunciation.

    Args:
        string1 (str): The first string.
        string2 (str): The second string.
        semantic_threshold (float): Semantic similarity threshold.
        fuzzy_threshold (int): Fuzzy matching threshold for text.
        phonetic_threshold (int): Phonetic similarity threshold.

    Returns:
        bool: True if the strings are semantically or phonetically equivalent, False otherwise.
    """
    for i in L:
      try:
          print('\n',i)
          # Translate both strings to English
          translated_string1 = translator.translate(string1, dest='en').text
          translated_string2 = translator.translate(i, dest='en').text
          print('New Original Title:', string1 ,'\nNew Translated Title:', translated_string1)
          print('Stored Original Title:', i ,'\nStored Translated Title:', i)

          # print(translated_string2)
          embedding1 = model.encode(string1)
          embedding2 = model.encode(i)
          embeddingt1 = model.encode(translated_string1)
          embeddingt2 = model.encode(translated_string2)

          semantic_similarity = util.cos_sim(embedding1, embedding2).item()
          semantic_similarity2 = util.cos_sim(embeddingt1, embeddingt2).item()


          # print(embedding1)
          # print(embedding2)

      except Exception as e:
          print(f"Translation Error: {e}")
          return False

      # Step 1: Phonetic similarity using fuzzy matching
      phonetic1 = metaphone(string1)
      phonetic2 = metaphone(i)
      phonetic3 = metaphone(translated_string1)
      phonetic4 = metaphone(translated_string2)
      print(phonetic1,phonetic2,phonetic3,phonetic4,)

      phonetic_similarity = fuzz.ratio(phonetic1, phonetic2)
      phonetic_similarity2= fuzz.ratio(phonetic3, phonetic4)



      print(f"Phonetic similarity score: {phonetic_similarity, phonetic_similarity2}")
      if phonetic_similarity >= phonetic_threshold:
          return True
          pass

      # Step 2: Fuzzy similarity for minor spelling variations
      fuzzy_similarity = fuzz.ratio(string1.lower(), i.lower())
      fuzzy_similarity2 = fuzz.ratio(translated_string1.lower(), translated_string2.lower())


      print(f"Fuzzy similarity score: {fuzzy_similarity, fuzzy_similarity2}")
      if fuzzy_similarity >= fuzzy_threshold:
          return True

      # Step 3: Semantic similarity for meaning
      print(f"Semantic similarity score: {semantic_similarity:.2f},{semantic_similarity2:.2f}")

      # Compare to semantic threshold
      return semantic_similarity >= semantic_threshold

L=['Dainik Jagran ','SAKHI','KHATIK','KHET KHALIHAN','SAMAJWADI']
# Example usage
string1 = "Jagran Dainik"
# string2 = "KRISAK JAGRAN"

if are_strings_equivalent(string1, L):
    print("The strings are phonetically or semantically equivalent!")
else:
    print("The strings are not equivalent.")
```
