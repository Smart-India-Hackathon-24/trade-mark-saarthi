```python
import pandas as pd

# Load the CSV file
csv_file = "../result_datasets/title_soundex.csv"  # Replace with your CSV file name
df = pd.read_csv(csv_file)

# Open the output text file in write mode
output_file = '../result_datasets/soundex_similar_titles.txt'

with open(output_file, 'w') as f:
    # Iterate over each column to group and sort duplicates
    for column in ['Soundex']:
        duplicates = df[df[column].duplicated(keep=False)]  # Rows with duplicates in the column
        grouped = duplicates.groupby(column).size().sort_values(ascending=False)  # Group and sort by count
        
        if not grouped.empty:  # Only display if there are duplicates
            f.write(f"Sorted duplicate groups for column '{column}':\n")
            f.write(f"{grouped}\n\n")
            
            # Display the full rows for each group of duplicates in descending order of their group size
            for value in grouped.index:  # Iterate over sorted group values
                group = duplicates[duplicates[column] == value]  # Filter the full rows of each group
                f.write(f"Group for '{column}' = {value} (Count: {grouped[value]}):\n")
                f.write(f"{group}\n\n")

print(f"Results have been saved to {output_file}")
```

```python
def check_string(input_str):
    # List of words to check for
    restricted_words = ["Police", "Crime", "Corruption", "CBI", "CID", "Army"]
    
    # Split the input string into individual words
    words = input_str.split()
    
    # Check if any word in the input string matches any restricted word
    for word in words:
        if word in restricted_words:
            return False
    
    # If none of the words matched, return True
    return True

# Example usage:
input_str = "This is a test string"
print(check_string(input_str))  # Should return True

input_str = "The Police arrived at the scene"
print(check_string(input_str))  # Should return False
```

```python
def remove_spaces_variants(s):
    # Generate all possible versions of the string with spaces removed at different positions
    variants = set()
    for i in range(len(s) + 1):
        # Create a variant with spaces removed up to index `i`
        variant = s[:i] + s[i:].replace(" ", "")
        variants.add(variant)
    return variants

def is_not_in_list(input_string, string_list):
    # Get all variants of the input string (spaces removed at any position)
    input_variants = remove_spaces_variants(input_string.replace(" ", "")) # First remove all spaces from input string
    
    # For each string in the list, generate the variants and check if there's a match
    for string in string_list:
        list_variants = remove_spaces_variants(string.replace(" ", ""))
        if any(variant in list_variants for variant in input_variants):
            return False
    return True
```

```python
# List of strings
string_list = ["hello world", "good morning", "python programming", "DAINIK JAGRAN RASHTRIYA SANSKARAN", "KRISHI JAGRAN"]

# Test input 1
input_string = "helloworld"
result = is_not_in_list(input_string, string_list)
print(result)  # Output: False (because "helloworld" matches "hello world" with spaces removed)

# Test input 2
input_string = "good night"
result = is_not_in_list(input_string, string_list)
print(result)  # Output: True (because "good night" does not match any string in the list)

# Test input 3
input_string = "python prog ramming"
result = is_not_in_list(input_string, string_list)
print(result)  # Output: False (because "python prog ramming" matches "python programming" with spaces removed)
```

```python
def is_valid_string(input_string, disallowed_prefixes, disallowed_suffixes):
    # Check each word in the input string
    words = input_string.split()
    
    for word in words:
        # Check for disallowed prefixes
        for prefix in disallowed_prefixes:
            if word.startswith(prefix):
                return False
        
        # Check for disallowed suffixes
        for suffix in disallowed_suffixes:
            if word.endswith(suffix):
                return False
    
    # If no disallowed prefix or suffix is found
    return True
```

```python
disallowed_prefixes = ["HINDUSTAN", "INDIA"]
disallowed_suffixes = ["TIMES", "DAY"]

print(is_valid_string("DESH BHAKT JAGRAN", disallowed_prefixes, disallowed_suffixes))  
print(is_valid_string("DESH BHAKT TIMES", disallowed_prefixes, disallowed_suffixes))  
print(is_valid_string("INDIA BHAKT TIMES", disallowed_prefixes, disallowed_suffixes))
```
